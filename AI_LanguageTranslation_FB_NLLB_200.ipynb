```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model_1 = AutoModelForSeq2SeqLM.from_pretrained("/content/model")
tokenizer_1 = AutoTokenizer.from_pretrained("/content/tokenizer")
```

```python
source_lang = 'eng_Latn'
target_lang = 'mar_Deva'
text = "RRR (subtitled onscreen as Roudram Ranam Rudhiram) is a 2022 Indian Telugu-language epic period action drama film directed by S. S. Rajamouli, who co-wrote the film with V. Vijayendra Prasad. It was produced by D. V. V. Danayya of DVV Entertainment. The film stars N. T. Rama Rao Jr., Ram Charan, Ajay Devgn, Alia Bhatt, Shriya Saran, Samuthirakani, Ray Stevenson, Alison Doody, and Olivia Morris. It revolves around the fictional versions of two Indian revolutionaries, Alluri Sitarama Raju (Charan) and Komaram Bheem (Rama Rao), their friendship, and their fight against the British Raj."

translator_eng_tam = pipeline('translation', model=model_1, tokenizer=tokenizer_1, src_lang=source_lang, tgt_lang=target_lang, max_length = 400)

result = translator_eng_tam(text)
print(result[0]['translation_text'])
```
